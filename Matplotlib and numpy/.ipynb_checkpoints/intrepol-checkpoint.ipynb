```python
import numpy as np
```

```python
from scipy.interpolate import interp1d
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, num = 11, endpoint = True)
```

```python
x
```

```python
y = np.cos(-x**2/9.0)
```

```python
y
```

```python
plt.plot(x, y, 'o')#145
```

```python
plt.plot(x, y, '-')
```

```python
# Линейная интерполяция. Извезтвные точки соединяются прямыми линиями
f = interp1d(x, y) 
```

```python
f(2.5) # Выводит промежуточное значение точк на графике
```

```python
xnew = np.linspace(0, 10, num = 41, endpoint = True)
plt.plot(x, y, 'o', xnew, f(xnew), '-')
plt.legend(['Таблица', 'Линейная интерпл.'], loc = 'best')
```

```python
#Кубическая сплайн-интерполяция
#Точки соединяются кубическими полиномами
f = interp1d(x, y, kind = 'cubic')
```

```python
f(2.5)
```

```python
f = interp1d(x, y, kind = 'cubic')
xnew = np.linspace(0, 10, num = 41, endpoint = True)
plt.plot(x, y, 'o', xnew, f(xnew), '-')
plt.legend(['Таблица', 'Линейная интерпл.'], loc = 'best')
```
