```python
import matplotlib as mpl
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
fig.savefig('my_figure.png') #сохраняет рисунок
```
