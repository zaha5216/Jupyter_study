```python
import numpy as np
import cv2

# Загрузка цветного изображения как полутонового  в виде табицы
img1 = cv2.imread('/Users/zaha/Downloads/data/1.png',0)
img2 = cv2.imread('/Users/zaha/Downloads/data/2.png',0)


MY_VIEW = cv2.WINDOW_NORMAL|cv2.WINDOW_AUTOSIZE
cv2.namedWindow('img1',MY_VIEW)
cv2.imshow('img1',img1)
cv2.namedWindow('img2',MY_VIEW)
cv2.imshow('img2',img2)

cv2.waitKey(0)
cv2.destroyAllWindiws()
```

```python
# запись изображения
import numpy as np
import cv2

img1 = cv2.imread('/Users/zaha/Downloads/data/1.png',0)
img2 = cv2.imread('/Users/zaha/Downloads/data/2.png',0)

cv2.imwrite('/Users/zaha/Documents/img1.png',img1,
           [cv2.IMWRITE_PNG_COMPRESSION, 1])
cv2.imwrite('/Users/zaha/Documents/img2.png',img2,
           [cv2.IMWRITE_PNG_COMPRESSION, 1])
cv2.destroyAllWindows()
```
