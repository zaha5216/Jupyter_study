```python
%matplotlib inline
from skimage.io import imread, imshow, imsave
img = imread('/Users/zaha/Documents/tigr-color .jpeg')
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img)
```

```python
r = img_f[:, :, 0] # Извлечение красного канала
```

```python
imshow(r)
```

```python
g = img_f[:, :, 1] # Извлечение зелёного канала
```

```python
imshow(g)
```

```python
b = img_f[:, :, 2] # Извлечение синиго канала
```

```python
imshow(b)
```

```python
from numpy import dstack
```

```python
img_combined = dstack((r, g, b,)) # Объединяем 3 канала
```

```python
imshow(img_combined)
```

```python
avg_gray = (r + g + b) / 3  # Усреднение 3 каналов
```

```python
imshow(avg_gray) 
```
