```python
# Наложение изображений 
img1 = cv2.imread('/Users/zaha/Documents/angel.jpg')
img2 = cv2.imread('/Users/zaha/Documents/budda.jpg')
imageShape = np.array([np.array(img1.shape),
                       np.array(img2.shape)])
imageSize = np.min(imageShape, axis = 0)
print(imageSize)
img1Crop = img1[0:imageSize[0],0:imageSize[1],:].copy()
img2Crop = img2[0:imageSize[0],0:imageSize[1],:].copy()

dst = cv2.addWeighted(img1Crop,0.3,img2Crop,0.7,0)

```
